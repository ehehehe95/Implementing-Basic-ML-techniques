```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
```

## Json to Dataframe

```python
with open("ko_nia_normal_squad_all.json", 'rt', encoding='UTF8') as file:
    json_original = json.load(file)
```

```python
json_original
```

```python
#dataframe to be created
#1. title : context
#2. title : questions & answer
df_normal_context = pd.DataFrame(columns = ["title","context"])
df_normal_qa = pd.DataFrame(columns = ["title", "question", "answer_start", "answer_text","classtype"])
c_idx = 0
q_idx = 0
for data in json_original["data"]:
    if c_idx == 1000:
        break
    title = data["title"]
    paragraphs = data["paragraphs"]
    df_normal_context.loc[c_idx] = [title,paragraphs[0]["context"]]
    c_idx += 1
    qas = paragraphs[0]["qas"]
    for qa in qas:
        df_normal_qa.loc[q_idx] = [title, qa["question"], qa["answers"][0]["answer_start"], qa["answers"][0]["text"], qa["classtype"]]
        q_idx += 1
```

```python
df_normal_context.to_csv("normal_context", mode='w')
df_normal_qa.to_csv("normal_qa", mode='w')
```

### remove stop words 
https://www.ranks.nl/stopwords/korean

```python
target = okt.nouns(df_normal_qa["question"][5])
stopword = pd.read_csv("stopwords.csv").단어.to_list()
```

```python
result = []
for word in target:
    if word not in stopword:
        result.append(word)
```

### tokenize & word2vec

https://ratsgo.github.io/natural%20language%20processing/2017/03/08/word2vec/


```python
tokenized_context = []
for context in df_normal_context["context"]:

    temp = okt.morphs(context, stem=True) # 토큰화
    temp = [word for word in temp if not word in stopword] # 불용어 제거
    tokenized_context.append(temp)
```

```python
print('max(len) :',max(len(l) for l in tokenized_context))
print('avg(len) :',sum(map(len, tokenized_context))/len(tokenized_context))
plt.hist([len(s) for s in tokenized_context], bins=50)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sentences = tokenized_context, size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
```
