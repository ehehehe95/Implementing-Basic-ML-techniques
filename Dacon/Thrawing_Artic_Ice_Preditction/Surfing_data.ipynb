# Read Data

## check whether other features matters

```python
def check_all_same(channel):
    c_train = train[:, :, :, channel]
    same = True
    for image in c_train:
        same = same and not(np.bitwise_xor(image,c_train[0]).any())
    return same
```

```python
for i in range(5):
    print("channel {} : {}".format(i, check_all_same(i)))
```
